```python
import pandas as pd
```

```python
data_type = '60'

raw_data = pd.read_csv(f'data_{data_type}.csv')
meta_data = pd.read_csv(f'meta_data_{data_type}.csv')

meta_data = meta_data.set_index('name', drop=True)
meta_data
```

```python
data = raw_data[raw_data.index.isin(meta_data.index)]
data = data.sort_index(key=lambda x: x.str.extract('t(\d+)').astype(int)[0])
data
```

```python
data['v'] = meta_data['v']
data['f'] = meta_data['f']
label = meta_data['t']
mtt_label = meta_data[['mtt24', 'mtt48']]
```

```python
data.to_csv(f'ml_ready_data_{data_type}.csv', index=False)
label.to_csv(f'temperature_labels_{data_type}.csv')
mtt_label.to_csv(f'mtt_labels_{data_type}.csv')
```
